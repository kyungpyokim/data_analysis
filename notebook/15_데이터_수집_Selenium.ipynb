# 9. 데이터 가져오기

## 1) 스크롤 아래로 내리기

```python
# 왜 내려야 하는지?
#

display_area_selector = "#yDmH0d > div.VfPpkd-Sx9Kwc.cC1eCc.UDxLd.PzCPDd.HQdjr.VfPpkd-Sx9Kwc-OWXEXe-FNFY6c > div.VfPpkd-wzTsW > div > div > div > div > div.fysCi.Vk3ZVd"
display_area = driver.find_element(By.CSS_SELECTOR, display_area_selector)

rep = 10
for i in range(rep):
    display_area.send_keys(Keys.END)
    display_area.wait
```

## 2) 스크롤 맨 처음으로 올라가기

```python
display_area.send_keys(Keys.HOME)
```

## 3) 댓글 추출하기

```python
review_selector = "#yDmH0d > div.VfPpkd-Sx9Kwc.cC1eCc.UDxLd.PzCPDd.HQdjr.VfPpkd-Sx9Kwc-OWXEXe-FNFY6c > div.VfPpkd-wzTsW > div > div > div > div > div.fysCi.Vk3ZVd > div > div:nth-child(2) > div"
reviews = driver.find_elements(By.CSS_SELECTOR, review_selector)
print(len(reviews))
```

### (1) 리뷰 하나만 살펴보기


```python
review = reviews[0]
review
```

```python
# 작성자
name = review.find_element(By.CLASS_NAME, "X5PpBb").text
print(name)
```

```python
# 평점 찾기
score = review.find_element(By.CLASS_NAME, "iXRFPc").get_attribute("aria-label")
```

```python
# 작성일
enroll_date = review.find_element(By.CLASS_NAME, "bp9Aid").text
enroll_date
```

```python
review_text = review.find_element(By.CLASS_NAME, "h3YV2d").text
print(review_text)
```

```python
# 찾은 데이터 합치기
row = {
    "name": name,
    "score": score,
    "enroll_date": enroll_date,
    "review_text": review_text,
}
print(row)
```

### (2) 모든 리뷰 데이터 모으기


```python
data_list = []

for i, review in enumerate(reviews):
    # 프린트 제어
    if i % 10 == 0:
        print(f"{i}번째 리뷰를 추출중입니다....")
    name = review.find_element(By.CLASS_NAME, "X5PpBb").text
    score = review.find_element(By.CLASS_NAME, "iXRFPc").get_attribute("aria-label")
    enroll_date = review.find_element(By.CLASS_NAME, "bp9Aid").text
    review_text = review.find_element(By.CLASS_NAME, "h3YV2d").text

    row = {
        "name": name,
        "score": score,
        "enroll_date": enroll_date,
        "review_text": review_text,
    }
    data_list.append(row)
```

```python
import pandas as pd

df = pd.DataFrame(data_list)
df.head()
```

```python
df.to_csv("./data/배달의민족_구글플레이스토어_댓글수집(2660128).csv")
```
