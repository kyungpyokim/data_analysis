## 3) 매개변수

```python
# 덧셈 계산기: 
## 입력: a, b 
## 실행: a + b 반환

## 함수이름: calc_add(a, b)
def calc_add(a, b):
    return a + b
```

```python
result = calc_add(a=3, b=4)
print(result)
```

```python
result = calc_add(b=4, a=3)
print(result)
```

```python
result = calc_add(3, 5) # 순서대로 입력
print(result)
```

```python
# 덧셈뺄셈 계산기
# 입력변수: a, b, operator
# 함수이름 calculator

def calculator(a, b, operator="+"):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    else:
        return "저는 덧셈, 뺄셈만 할 수 있어요."
```

```python
result = calculator(7, 2)
print(result)
```

```python
## 변형해보기: 사칙연산 계산기 만들기 (나눗셈 조건 추가해보기)
def calculator(a, b, operator="+"):
    if operator == "+":
        return a + b
    elif operator == "-":
        return a - b
    elif operator == "*":
        return a * b
    elif operator == "/":
        if b == 0:
            return "0으로 나눌 수 없습니다."
        return a / b
    else:
        return f"알 수 없는 연산자 입니다. ({operator})"
```

```python
calculator(1, 0, "/")
```

### 심화

```python
# 사용자 정보를 받는 함수
# 입력이 자유롭다. name만 입력해도 되고, name, age만 입력해도 작동하도록 한다. 
def user_info(*args):
    print(args)
```

```python
user_info('홍길동')
user_info('홍길동', 23)
user_info('홍길동', 23, 188.2)
```

```python
def user_info(**kwargs):
    print(kwargs)
```

```python
user_info(name="홍길동", age=23, height="188.2")
```

## 4) lambda 함수

```python
def calc_add(a, b):
    return a + b
```

```python
calc_add(2, 5)
```

```python
calc_add = lambda a, b: a + b

calc_add(5, 8)
```

## 2) 파일 생성/읽기/추가

### 생성

```python
# mode: w(write), r(read), a(append)
with open("./data/mytext.txt", mode="w", encoding="utf-8") as f:
    f.write("안녕")
```

### 읽기

```python
with open("./data/mytext.txt", mode="r", encoding="utf-8") as f:
    data = f.read()
    print(data)
```

### 추가

```python
with open("./data/mytext.txt", mode="a", encoding="utf-8") as f:
    f.write("반가워")
```

```python
aespa = ["카리나", "윈터", "지젤", "닝닝"]

# 카리나
# 윈터
# 지젤
# 닝닝

with open("./data/aespa.txt", mode="w", encoding="utf-8") as f:
    for member in aespa:
        f.write(member + "\n") # \n: 줄바꿈
```
