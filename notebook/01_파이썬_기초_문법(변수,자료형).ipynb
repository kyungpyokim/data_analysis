````python
## 수학 문제 
## 100 / 30 몫 3 나머지 10
## 30개씩 3페이지가 꽉 차고 다음 페이지에 10개의 게시물이 올라와야 한다. 
## 결론 4페이지를 만들어야 한다. 

# 1. 변수 정의
posts = 100
posts_per_page = 30

# 2. 연산 설명
pages = (posts // posts_per_page) + 1
remainder = posts % posts_per_page
print("100 나누기 30의 나머지", remainder)
print("나머지가 있기 때문에 몫에 1을 더해서", pages)

print("="*100)
## 만약에 120개의 게시물을 한 페이지에 30개씩 보여주려면 총 몇 페이지가 필요할까요
# 1. 변수 정의
posts = 120
posts_per_page = 30

# 2. 연산 설명
pages = (posts // posts_per_page)
remainder = posts % posts_per_page
print("120 나누기 30의 나머지", remainder)
print("나머지가 없기 때문에 그대로", pages)

## 자동화 한다. 
## 1. 조건 : 나머지가 0이니?
## 맞다 -> + 1 하지 않는다.
## 아니다 -> + 1을 해야 한다.
````

* 미세먼지(`pm10`)키 추가하기 (값: 34)
* 'forecast' 키에 ['Rainy', 'Rainy'] 추가하기

```
weather_data = {
    'city': 'Seoul',
    'temperature': 22.5,  # 섭씨
    'humidity': 60,       # 상대 습도(%)
    'forecast': ['Sunny', 'Cloudy', 'Rainy']  # 3일 예보
}
```

````python
weather_data = {
    'city': 'Seoul',
    'temperature': 22.5,  # 섭씨
    'humidity': 60,       # 상대 습도(%)
    'forecast': ['Sunny', 'Cloudy', 'Rainy']  # 3일 예보
}
print(weather_data)
print("=" * 100)
print("city 키를 Busan으로 수정")
print(weather_data["city"])
weather_data["city"] = "Busan"
print(weather_data)
print("=" * 100)
print("pm10 키를 값을 34로 추가하기")
weather_data["pm10"] = 34
print(weather_data)
print("=" * 100)
print("forecast에 ['Rainy', 'Rainy'] 추가하기")
print(weather_data["forecast"].extend(["Rainy", "Rainy"]))
print(weather_data)
````

### (3) 관련 메서드

````python
student = {
    "name": "홍길동",
    "age": 28,
    "height": 177.6,
    "score": [76, 88, 96, 76]
}
print(student)
````

#### 조회

````python
# 딕셔너리 안에 있는 모든 키들만 추출하기 
print(student.keys(), type(student.keys()))

key_list = list(student.keys())
print(key_list)
````

````python
# 딕셔너리 안에 있는 모든 값들만 추출하기 
print(student.values(), type(student.values()))

value_list = list(student.values())
print(value_list)
````

````python
# 딕셔너리 안에 있는 모든 (키, 값) 쌍 추출하기 
print(student.items(), type(student.items()))

item_list = list(student.items())
print(item_list)
````

````python
# 조건 파악: 내가 찾고자 하는 키가 있는가?
print("height" in student)
print("pressure" in student)
````

````python
# 값 조회 
print(student["height"])
print(student["pressure"])
````

````python
# 값 조회 - get
print(student.get("pressure", "없어요"))
````

#### 삭제

````python
student = {
    "name": "홍길동",
    "age": 28,
    "height": 177.6,
    "score": [76, 88, 96, 76]
}
print(student)
````

````python
# 삭제 - del 
del student["age"]
print(student)
````

````python
# 전체 삭제 
student.clear()
print(student)
````

### (2) 요소 추가/제거

#### 추가

````python
myset = {1, 2, 3, 4, 5}
print(myset)

# 하나 추가 
myset.add(6)
print(myset)

myset.add(2)
print(myset)

# 여러 개 추가
myset.update([3, 7, 8])
print(myset)
````

#### 제거

````python
myset = {1, 2, 3, 4, 5}

# 제거 - remove
myset.remove(2)
print(myset)

myset.remove(6)
print(myset)
````

````python
myset = {1, 2, 3, 4, 5}

# 제거 - discard
myset.discard(2)
print(myset)

myset.discard(6)
print(myset)
````

### (3) 연산

````python
yoga = {"A", "C", "D", "F", "G"}
health = {"A", "B", "F", "H"}

# 교집합 (AND)
set_inter = yoga & health
print(set_inter)

set_inter = yoga.intersection(health)
print(set_inter)

set_inter = health.intersection(yoga)
print(set_inter)
````

````python
yoga = {"A", "C", "D", "F", "G"}
health = {"A", "B", "F", "H"}

# 합집합 (OR)
set_union = yoga | health
print(set_union)

set_union = yoga.union(health)
print(set_union)

set_union = health.union(yoga)
print(set_union)
````

````python
yoga = {"A", "C", "D", "F", "G"}
health = {"A", "B", "F", "H"}

# 차집합 (요가-헬스 : 요가만 하는 사람, 헬스-요가: 헬스만 하는 사람)
## 요가-헬스 : 요가만 하는 사람
set_diff = yoga - health 
print(set_diff)

set_diff = yoga.difference(health)
print(set_diff)

## 헬스-요가: 헬스만 하는 사람
set_diff = health - yoga 
print(set_diff)

set_diff = health.difference(yoga)
print(set_diff)
````
