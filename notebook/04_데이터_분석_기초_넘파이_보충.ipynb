```python
mylist1 = [1, 2, 3]
mylist2 = [4, 5, 6]

print(mylist1 + mylist2)
# (예상답안: [5, 7, 9]) 각 요소별 합을 구하세요. (반복문)
# 변수 정의 
total = []
# 반복 변수 i
for i in range(len(mylist1)):
    print(f"{mylist1[i]} + {mylist2[i]} = {mylist1[i] + mylist2[i]}" )
    total.append(mylist1[i] + mylist2[i])

print(total)
```

```python
# 중간고사 점수(국어, 수학, 영어, 과학)
짱구 = [90, 87, 68, 91]
철수 = [88, 85, 96, 100]
유리 = [65, 78, 94, 86]
# 과목별로 떡잎유치원 평균을 알고 싶다. 
# 반복문
mean_list = []
for i in range(len(짱구)):
    mean_result = 짱구[i] + 철수[i] + 유리[i] / 3
    print(f"{짱구[i]}+{철수[i]}+{유리[i]} / 3")
    mean_list.append(mean_result)

print(mean_list)
```

## 2) 패턴이 있는 배열 생성

```python
import numpy as np 

# 모든 값이 0인 배열을 생성(np.zeros(shape))
data1 = np.zeros((2, 3))
print(data1)

# 모든 값이 1인 배열을 생성(np.ones(shape))
data2 = np.ones((2, 2, 2))
print(data2)

# 모든 값이 내가 원하는 값인 배열을 생성(np.full(shape, 내가 원하는 값))
data3 = np.full((4, 2), 9)
print(data3)
```

```python
# 연속된 정수 배열 생성 np.arange(start, end, step)
# start 부터 end 앞까지 step 간격으로 배열 생성
data4 = np.arange(0, 10, 2)
print(data4)
```

```python
# 정해진 범위 동일한 간격 배열 생성 np.linspace(start, end, n)
# start부터 end까지 동일한 간격의 n개의 데이터 배열을 생성
data5 = np.linspace(0, 1, 5)
print(data5)
```

## 3) 난수 배열 생성

```python
import numpy as np 

np.random.seed(123)

# 0과 1사이의 실수 랜덤 배열 생성 np.random.rand(axis0, axis1, axis2)
random_arr1 = np.random.rand(2, 3)
print(random_arr1)

# 정규분포 (평균 0, 표준편차 1) np.random.randn(axis0, axis1, axis2)
# 평균이 0이고 표준편차가 1인 랜덤 배열 생성
random_arr2 = np.random.randn(2, 3)
print(random_arr2)

# 특정 범위 내 정수 랜덤 배열 생성 np.random.randint(start, end, shape)
random_arr3 = np.random.randint(0, 10, (2, 3))
print(random_arr3) 
```

# 4. 배열 변환

```python
import numpy as np 

data = np.arange(0, 15)
print(data)
print(data.size)

# shape를 변환 (3, 5) 배열.reshape()
data_reshape = data.reshape(3, 5)
print(data_reshape)

# 전치행렬 (행/열 전환) 배열.transpose()
data_transpose = data_reshape.transpose()
print(data_transpose)
data_transpose2 = data_reshape.T
print(data_transpose2)

# n차원 배열을 1차원으로 변환 배열.flatten()
data_flatten = data_transpose.flatten()
print(data_flatten)
```

# 5. 배열 연산

```python
# 연산: 같은 위치에 있는 요소끼리
# 연산할 때에는 shape가 같아야 한다.
data1 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
data2 = np.array([
    [10, 20, 30],
    [40, 50, 60]
])
print(data1)
print(data2)
```

```python
print("배열의 덧셈")
print(data1 + data2)
print("="*30)
print("배열의 뺄셈")
print(data1 - data2)
print("="*30)
print("배열의 곱셈")
print(data1 * data2)
print("="*30)
print("배열의 나눗셈")
print(data1 / data2)
print("="*30)
print("배열의 거듭제곱")
print(data2 ** data1)
print("="*30)
```

```python
# data1 배열에 모든 요소에 10을 곱하고 싶다. 
print(data1)
ten_data = np.full((2, 3), 10)
print(ten_data)

print(data1 * ten_data)
print(data1 * 10)
```

```python
# 통계
data1 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
print(data1)
print(np.sum(data1))    # 전체 합
print(np.mean(data1))   # 전체 평균
print(np.max(data1))    # 전체 요소 중 최댓값
print(np.min(data1))    # 전체 요소 중 최솟값
print(np.var(data1))    # 전체 분산
print(np.std(data1))    # 전체 표준편차
```

```python
data1 = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print(data1)
print("axis=0 요약")
print(np.sum(data1, axis=0))
print("axis=1 요약")
print(np.sum(data1, axis=1))
```

```python
data2 = np.array([
    [
        [255, 0, 0],
        [0, 255, 0]
    ],
    [
        [0, 0, 255],
        [255, 255, 0]
    ]
])
print(data2)
print("="*30)
print("axis=0 요약")
print(np.sum(data2, axis=0))
print("="*30)
print("axis=1 요약")
print(np.sum(data2, axis=1))
print("="*30)
print("axis=2 요약")
print(np.sum(data2, axis=2))
```
