#### 문자열 대체(replace)

````python
gender = "mfmfmfmfmfmffffmmmmmMMMMFFFFMFFMM"

print(gender.replace("m", "M"))
print(gender.replace("f", "F"))
````

````python
gender = "mfmfmfmfmfmffffmmmmmMMMMFFFFMFFMM"
print(gender)
gender2 = gender.replace("m", "M")
print(gender2)
gender3 = gender2.replace("f", "F")
print(gender3)
````

#### 대/소문자 변환(lower, upper)

````python
gender = "mfmfmfmfmfmffffmmmmmMMMMFFFFMFFMM"
print(gender.lower())
print(gender.upper())
````

#### 공백 제거(lstrip/rstrip/strip)

````python
name1 = "홍길동"
name2 = "홍길동 " 
name3 = "홍길동    "

print(name1, len(name1))
print(name2, len(name2))
print(name3, len(name3))
````

````python
name = "    홍길동             "
print(f"|{name}|")
print(f"|{name.strip()}|")
print(f"|{name.lstrip()}|")
print(f"|{name.rstrip()}|")
````
