````python
score_list = [55, 80, 62, 45, 90]

# 반복 변수 정의: score 
# 반복 출력: 조건 만족하면 "합격", 아니면 "불합격"
# 조건: score가 60점 이상, score >= 60
for score in score_list:
    if score >= 60:
        print(f"score: {score}, 합격")
    else:
        print(f"score: {score}, 불합격")
````

# 11. 숫자를 1부터 출력하다가 10을 넘으면 반복을 즉시 멈추세요(break).


````python
# 반복 변수 정의: x
# 반복 출력: 1부터 차례대로 출력
# 조건: 10을 출력한 뒤 멈춤, x == 10 break
x = 0

while True:
    x += 1
    print(x)
    if x == 10:
        break
````

````python
# 반복 변수 정의: x
# 반복 출력: 1부터 21 앞까지 출력
# 조건1: 짝수, x % 2 == 0
# 조건2: 3의 배수가 아닌 것, x % 3 != 0
# 조건 결합: and
for x in range(1, 21):
    if x % 2 == 0 and x % 3 != 0:
        print(x)
````

# 16. [구형 연락처 추출] 
```
users = [
    {"name": "A", "phone": "011-1234-5678"}, 
    {"name": "B", "phone": "010-5678-4321"}, 
    {"name": "C", "phone": "011-9999-1231"}
]
```
전화번호(phone)가 "011"로 시작하는 고객의 이름만 출력하세요. (문자열 .startswith() 활용해보기)

````python
# startswith(특정 문자)란? 특정 문자로 시작하는지 확인
# endswith(특정 문자)란? 특정 문자로 끝나는지 확인
text = "2025년 1월 16일 금요일"

print(text.startswith("2025"))
print(text.startswith("2020"))
print(text.endswith("금요일"))
print(text.endswith("요일"))
````

````python
users = [
    {"name": "A", "phone": "011-1234-5678"}, 
    {"name": "B", "phone": "010-5678-4321"}, 
    {"name": "C", "phone": "011-9999-1231"}
]

# 반복 변수 정의: user_dict 
# 반복 출력: 조건에 만족하면 user_dict의 이름 출력, user_dict["name"]
# 조건: 키가 phone인 값에서 011로 시작한다, user_dict["phone"].startswith("011")
for user_dict in users:
    if user_dict["phone"].startswith("011"):
        print(user_dict["name"])
````


# 17. [재고 금액 합산]
```
stock = {"apple": 5, "banana": 0, "cherry": 10}
prices = {"apple": 1000, "banana": 500, "cherry": 2000}
```
현재 가지고 있는 모든 과일의 총 재고 가치(수량 * 가격)를 계산하세요.


````python
# 변수 정의 
stock = {"apple": 5, "banana": 0, "cherry": 10}
prices = {"apple": 1000, "banana": 500, "cherry": 2000}
total_value = 0

# 반복할 대상: stock의 키 또는 prices의 키
# 반복 변수 정의: key
# 반복 행위: stock의 key에 대한 값과 prices의 key에 대한 값을 곱한다. 
#          그리고 total_value에 더한다.
#          total_value += stock[key] * prices[key]
for key in stock:
    print(f"stock x key = {stock[key] * prices[key]}")
    total_value += stock[key] * prices[key]

print(total_value)
````

# 18. [이메일 도메인 필터링]
```
emails = ["test@naver.com", "user@gmail.com", "admin@naver.com", "hello@daum.net"]
```
네이버(naver.com) 계정을 사용하는 이메일만 추출하여 리스트를 만드세요.


````python
# 변수 정의
emails = ["test@naver.com", "user@gmail.com", "admin@naver.com", "hello@daum.net"]
naver_emails = []

# 반복 변수 정의: email
# 반복 행위: 조건을 만족하면 naver_emails에 넣는다, naver_emails.append(email)
# 조건: naver.com으로 끝나는가, email.endswith("naver.com")
for email in emails:
    if email.endswith("naver.com"):
        naver_emails.append(email)

print(naver_emails)
````

# 19. [장바구니 결제]
```
cart = [
    {"item": "T-shirt", "price": 20000, "amount": 2}, 
    {"item": "Pants", "price": 30000, "amount": 1}
]
```
전체 상품의 총 결제 금액을 반복문을 통해 구하세요.


````python
# 변수 정의
cart = [
    {"item": "T-shirt", "price": 20000, "amount": 2}, 
    {"item": "Pants", "price": 30000, "amount": 1}
]
total_price = 0

# 반복 변수 정의: product_dict
# 반복 행위: product_dict의 price와 amount를 곱한 후, price에 더한다.
#          price = product_dict["price"] * product_dict["amount"]
#          total_price += price 
for product_dict in cart:
    price = product_dict["price"] * product_dict["amount"]
    total_price += price 

print(total_price)
````

# 20. [데이터 클렌징] (isinstance() 참고) 
```
raw_data = [85, None, 92, "Missing", 78, None]
```
숫자(int) 데이터만 골라내어 새로운 리스트에 담고 그 평균을 구하세요.

````python
# isinstance(확인할 변수, 타입): 변수의 자료형을 확인하기 위한 함수
x = "1"

print(isinstance(x, str))
print(isinstance(x, int))
````

````python
raw_data = [85, None, 92, "Missing", 78, None]

# 반복 변수 정의: x
# 반복 출력: 조건에 맞는지 아닌지를 출력
# 조건: x의 type이 int인가, isinstance(x, int)
for x in raw_data:
    if isinstance(x, int):
        print(f"{x}의 자료형은 int입니다.")
    else:
        print(f"{x}의 자료형은 {type(x)}입니다.")
````
