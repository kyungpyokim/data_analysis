* https://www.data.go.kr/data/15077871/openapi.do 법정동코드 API로 불러와서 데이터로 저장하기


```python
base_url = "http://apis.data.go.kr/1741000/StanReginCd/getStanReginCdList"
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
import os
from urllib.parse import unquote

params = {
    "serviceKey": unquote(os.getenv("PUBLIC_DATA_API_KEY")),
    "pageNo": 1,
    "numOfRows": 3,
    "type": "json",
    "locatadd_nm": "서울특별시",
}

print(params)
```

```python
serviceKey = os.getenv("PUBLIC_DATA_API_KEY")
print(serviceKey)
pageNo = (1,)
numOfRows = 3
type = "json"
locatadd_nm = "서울특별시"

url = f"{base_url}?ServiceKey={serviceKey}&pageNo={pageNo}&numOfRows={numOfRows}&type={type}&locatadd_nm={locatadd_nm}"
```

```python
import requests as req

res = req.get(base_url, params=params)
print(res.status_code)

res = req.get(url)
print(res.status_code)

# import requests

# url = 'http://apis.data.go.kr/1741000/StanReginCd/getStanReginCdList'
# params ={'serviceKey' : '서비스키', 'pageNo' : '1', 'numOfRows' : '3', 'type' : 'xml', 'locatadd_nm' : '서울특별시' }

# response = requests.get(url, params=params)
# print(response.content)
```
