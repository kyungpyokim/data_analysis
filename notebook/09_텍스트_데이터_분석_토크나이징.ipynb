```python
# 0. ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¥¼ ë§Œë“ ë‹¤. word_list, stopwords
word_list = []
stopwords = ["ë°°ë¯¼", "ë°°ë‹¬"]

# 1. ë°ì´í„°í”„ë ˆì„ì—ì„œ text ê°’ì„ í•˜ë‚˜ì”© ë½‘ëŠ”ë‹¤. -> ë¬¸ì¥ sent
for sent in df_new["text"]:
    # 2. sentì—ì„œ í•„ìš”ì—†ëŠ” ë¬¸ì(íŠ¹ìˆ˜ë¬¸ì, ì´ëª¨ì§€ ë“±)ë¥¼ ì—†ì•¤ë‹¤. (íŒ¨í„´: [^0-9a-zA-Zê°€-í£\s])
    clean_sent = re.sub("[^0-9a-zA-Zê°€-í£\s]", "", sent)
    # 3. í˜•íƒœì†Œ ë¶„ì„ê¸°ë¡œ ë¬¸ì¥ì„ ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ë¡œ ë½‘ëŠ”ë‹¤.(ì¡°ê±´: Noun, ë‹¨ì–´ê¸¸ì´ê°€ 1ë³´ë‹¤ í°ê²ƒ, stopwordsì— ì—†ëŠ”ê²ƒ) -> result
    ## 3-1. (ë‹¨ì–´, í’ˆì‚¬) ìŒì˜ ë¦¬ìŠ¤íŠ¸ë¡œ ê²°ê³¼ë¥¼ ì¶œë ¥í•œë‹¤. 
    result = okt.pos(clean_sent)
    ## 3-2. í•˜ë‚˜ì”© ë½‘ì•„ì„œ í’ˆì‚¬ê°€ Nounì¸ì§€ í™•ì¸í•œë‹¤.(ë°˜ë³µ)
    sub_list = []
    for res in result:
        word = res[0]
        pos = res[1]
        ## 3-3. wordê°€ stopwordsì— ìˆìœ¼ë©´ ê±´ë„ˆë›´ë‹¤.
        if word in stopwords:
            continue
        ## 3-4. í’ˆì‚¬ê°€ Nounì´ê³  word ê¸¸ì´ê°€ 1ë³´ë‹¤ í° ê²ƒì„ sub listì— ë‹´ëŠ”ë‹¤. 
        if pos == "Noun" and len(word) > 1:
            sub_list.append(word)
    print(sub_list)
    # 4. word_listì— ì¡°ê±´ì— ë”°ë¼ ì¶”ì¶œí•œ result ìš”ì†Œë“¤ì„ ì¶”ê°€í•œë‹¤. 
    word_list.extend(sub_list)
    # print(f"{sent[:40]}")
    # print(f"{clean_sent[:40]}")
    # print(result)
    print("="*100)
```

```python
# ë¬´ì¡°ê±´ ë§ì´ ë°œìƒí•œ ë‹¨ì–´ë“¤ë§Œ ì˜ë¯¸ê°€ ìˆì„ê¹Œ? --> ë°ì´í„° ë§ˆì´ë‹: ìˆ¨ì–´ì ¸ ìˆëŠ” ê²ƒì„ ì°¾ì•„ëƒ„, ì˜ì™¸ì˜ ë‹¨ì–´ í‚¤ì›Œë“œë¥¼ ì°¾ì•„ë³´ì.
# ì›Œë“œí´ë¼ìš°ë“œë§Œ ë³´ê³  í…ìŠ¤íŠ¸ ë¶„ì„ì„ í•  ìˆ˜ ìˆì„ê¹Œ? --> ì‹¤ì œ ë°ì´í„°ë¥¼ ì°¾ì•„ë³´ê³  íŠ¹ì§• íŒŒì•…í•˜ê¸°
```

```python
# ë°ì´í„°í”„ë ˆì„ì—ì„œ íŠ¹ì • ë‹¨ì–´ê°€ í¬í•¨ëœ ë°ì´í„°ë§Œ ì¶œë ¥í•˜ë ¤ë©´ ì–´ë–»ê²Œ í•´ì•¼ í• ê¹Œìš”? 
# ì „ì²˜ë¦¬ ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°
import pandas as pd 

df_new = pd.read_csv(
    "./data/appreply2.csv", # "/content/appreply2.csv",
    index_col=0
)
df_new.head(2)
```

```python
keyword = "ë¡œê·¸ì¸"

# df_new["text"].str.contains(keyword) # ì¡°ê±´
login_df = df_new.loc[ df_new["text"].str.contains(keyword) , : ]
login_df.head(2)
```

```python
from konlpy.tag import Okt

okt = Okt()
```

```python
# login_dfì˜ textì—ì„œ ë‹¨ì–´ë“¤ì„ ë½‘ì•„ë‚´ì„œ ì›Œë“œí´ë¼ìš°ë“œë¥¼ ê·¸ë ¤ë³´ê¸°
import re 

# 0. ë¹ˆ ë¦¬ìŠ¤íŠ¸ë¥¼ ë§Œë“ ë‹¤. word_list, stopwords
word_list = []
stopwords = ["ë°°ë¯¼", "ë°°ë‹¬", "ë¡œê·¸ì¸"]

# 1. ë°ì´í„°í”„ë ˆì„ì—ì„œ text ê°’ì„ í•˜ë‚˜ì”© ë½‘ëŠ”ë‹¤. -> ë¬¸ì¥ sent
for sent in login_df["text"]:
    # 2. sentì—ì„œ í•„ìš”ì—†ëŠ” ë¬¸ì(íŠ¹ìˆ˜ë¬¸ì, ì´ëª¨ì§€ ë“±)ë¥¼ ì—†ì•¤ë‹¤. (íŒ¨í„´: [^0-9a-zA-Zê°€-í£\s])
    clean_sent = re.sub("[^0-9a-zA-Zê°€-í£\s]", "", sent)
    # 3. í˜•íƒœì†Œ ë¶„ì„ê¸°ë¡œ ë¬¸ì¥ì„ ë‹¨ì–´ ë¦¬ìŠ¤íŠ¸ë¡œ ë½‘ëŠ”ë‹¤.(ì¡°ê±´: Noun, ë‹¨ì–´ê¸¸ì´ê°€ 1ë³´ë‹¤ í°ê²ƒ, stopwordsì— ì—†ëŠ”ê²ƒ) -> result
    ## 3-1. (ë‹¨ì–´, í’ˆì‚¬) ìŒì˜ ë¦¬ìŠ¤íŠ¸ë¡œ ê²°ê³¼ë¥¼ ì¶œë ¥í•œë‹¤. 
    result = okt.pos(clean_sent)
    ## 3-2. í•˜ë‚˜ì”© ë½‘ì•„ì„œ í’ˆì‚¬ê°€ Nounì¸ì§€ í™•ì¸í•œë‹¤.(ë°˜ë³µ)
    sub_list = []
    for res in result:
        word = res[0]
        pos = res[1]
        ## 3-3. wordê°€ stopwordsì— ìˆìœ¼ë©´ ê±´ë„ˆë›´ë‹¤.
        if word in stopwords:
            continue
        ## 3-4. í’ˆì‚¬ê°€ Nounì´ê³  word ê¸¸ì´ê°€ 1ë³´ë‹¤ í° ê²ƒì„ sub listì— ë‹´ëŠ”ë‹¤. 
        if pos == "Noun" and len(word) > 1:
            sub_list.append(word)
    print(sub_list)
    # 4. word_listì— ì¡°ê±´ì— ë”°ë¼ ì¶”ì¶œí•œ result ìš”ì†Œë“¤ì„ ì¶”ê°€í•œë‹¤. 
    word_list.extend(sub_list)
    # print(f"{sent[:40]}")
    # print(f"{clean_sent[:40]}")
    # print(result)
    print("="*100)
```

```python
from collections import Counter 

counter = Counter(word_list)
```

```python
from wordcloud import WordCloud 
import matplotlib.pyplot as plt
import koreanize_matplotlib

wc = WordCloud(
    font_path = "C:\Windows\Fonts\malgun.ttf", # "/usr/share/fonts/truetype/nanum/NanumGothic.ttf"
    background_color="white",
    width=800,
    height=400
)

wc.generate_from_frequencies(counter)

plt.figure(figsize=(5,5))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("ë¡œê·¸ì¸ì´ ì–¸ê¸‰ëœ ë°°ë‹¬ì˜ ë¯¼ì¡± ì›Œë“œ í´ë¼ìš°ë“œ(ëª…ì‚¬)", fontsize=15)
plt.show()
```
