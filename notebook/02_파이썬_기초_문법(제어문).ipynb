### (2) continue

````python
# while True:, count = 5일 때 멈추기, 짝수일 때는 출력하지 말기
# 변수 정의
count = 0

# 조건: count == 5, count % 2 == 0
while True:
    count += 1

    # 조건 1
    if count % 2 == 0:
        continue

    print(count)

    # 조건 2
    if count == 5:
        break

# 현재 count = 0 -> count = 1 -> 조건1 확인: False -> print(1) -> 조건2 확인: False
# 현재 count = 1 -> count = 2 -> 조건1 확인: True
# 현재 count = 2 -> count = 3 -> 조건1 확인: False -> print(3) -> 조건2 확인: False
# 현재 count = 3 -> count = 4 -> 조건1 확인: True
# 현재 count = 4 -> count = 5 -> 조건1 확인: False -> print(5) -> 조건2 확인: True
````

````python
# 예제: 1부터 10까지 더하기
## 계획세우기
## 변수 정의 : result = 0, num = 0
## num은 반복이 될 때마다 1씩 늘어난다. 
## 반복이 될 때마다 result에 num을 계속 더한다.

## 설계
num = 0
result = 0

while True:
    num += 1
    print(f"number: {num}")
    print(f"result + number: {result} + {num}")
    result += num
    print(f"현재까지의 합: {result}")
    print("-" * 30)

    if num == 10:
        break

## 변형해보기: 1부터 10까지 짝수만 더하기 / 1부터 10까지 3의 배수만 더하기
````

````python
## 변형해보기: 1부터 10까지 짝수만 더하기 
# 변수 정의
num = 0
total = 0 
# 설계
while True:
    num += 1
    print(f"num: {num}")
    if num % 2 == 0:
        print(f"total = {total} + {num}")
        total += num
        print(f"total = {total}")
    else:
        print("더하지 않습니다.")

    if num == 10:
        break

    print("="*30)
````

````python
## 변형해보기: 1부터 10까지 3의 배수만 더하기
# 변수 정의
num = 0
total = 0 
# 설계
while True:
    num += 1
    print(f"num: {num}")
    if num % 3 == 0:
        print(f"total = {total} + {num}")
        total += num
        print(f"total = {total}")
    else:
        print("더하지 않습니다.")

    if num == 10:
        break

    print("="*30)
````

````python
# 예제: 비밀번호가 맞을 때까지 입력하기
## 계획세우기
## 변수 정의 password = wanted
## 반복: 입력(input()), 조건: if password == "wanted" 반복 끝

## 설계
password = "wanted"

while True:
    input_text = input("비밀번호를 입력하세요")

    if input_text == password:
        print("로그인 되었습니다.")
        break
    else:
        print("비밀번호가 틀렸습니다.")

## 변형해보기: 아이디, 비밀번호를 모두 설정해서 맞는지 확인하기 / 3회 이상 틀리면 종료
````

````python
# 총 매출 계산
orders = [
    {"customer": "Kim", "amount": 120000},
    {"customer": "Lee", "amount": 85000},
    {"customer": "Park", "amount": 99000},
]
total = 0 # 담을 변수 생성

for order in orders:
    print(order) # 딕셔너리
    print(order['amount'])
    total += order['amount']

print("="*30)
print(total)
````

## 4) List Comprehension

````python
# 빈 리스트에 1부터 10까지 숫자를 넣어보자. 
num_list = []
print(num_list)

for i in range(1, 11):
    num_list.append(i)

print(num_list)
````

````python
num_list = [i for i in range(1, 11)]
print(num_list)
````

````python
# 빈 리스트에 1부터 10까지 짝수를 넣어보자. 
num_list = []
print(num_list)

for i in range(1, 11):
    if i % 2 == 0:
        num_list.append(i)

print(num_list)
````

````python
num_list = [i for i in range(1, 11) if i % 2 == 0]
print(num_list)
````

````python
# 빈 리스트에 1부터 10까지 짝수면 "짝수" 아니면 "홀수"로 넣어보자.
num_list = []
print(num_list)

for i in range(1, 11):
    if i % 2 == 0:
        num_list.append("짝수")
    else:
        num_list.append("홀수")

print(num_list)
````

````python
num_list = ["짝수" if i % 2 == 0 else "홀수" for i in range(1, 11)]
print(num_list)
````
