# 1. 미니 실습

```python
html = """
<html>
  <body>
    <div id="main" class="content">
      <h1>첫 번째 헤더</h1>
      <h1 class="header">두 번째 헤더</h1>
      <p class="description">설명입니다.</p>
      <a href="https://example1.com" class="link">링크 1</a>
      <a href="https://example2.com" class="link">링크 2</a>
      <a href="https://example3.com" class="external">링크 3</a>
    </div>
    <div id="main2" class="content">
      <div class="items">
        <div>
            <img class="thumbnail" src="/data/picture1.png" />
            <div class="title">타이틀1</div>
            <div class="description">설명1</div>
        </div>
        <div>
            <img class="thumbnail" src="/data/picture2.png" />
            <div class="title">타이틀2</div>
            <div class="description">설명2</div>
        </div>
        <div>
            <img class="thumbnail" src="/data/picture3.png" />
            <div class="title test">타이틀3</div>
            <div class="description">설명3</div>
        </div>
      </div>
    </div>
  </body>
</html>
"""
```

## 1) HTML 파싱 준비

```python
# 요청을 해서 response.text를 html이라는 변수에 담았다고 가정하자. 
# HTML 파싱할 준비 
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, 'html.parser')
soup
```

## 2) 추출해보기

### soup.find("태그이름", class_=클래스이름, id=아이디이름)

```python
# 조건에 만족하는 맨 처음 요소 1개만 가져온다. 
# class이름이 title이라고 써진 요소를 가져와보자. 
soup.find("div", class_="title")
```

```python
# 텍스트 추출
soup.find("div", class_="title").text
```

### soup.find_all("태그이름", class_=클래스이름, id=아이디이름)

```python
# 조건에 만족하는 모든 요소들을 가져온다.
# class이름이 title이라고 써진 요소를 가져와보자. 
soup.find_all("div", class_="title")
```

```python
# 텍스트 추출: 반복문 이용
for element in soup.find_all("div", class_="title"):
    print(element.text)
```

### 속성 찾기

```python
# img 태그를 추출해주세요
print(soup.find("img"))
print(soup.find("img").attrs) # 속성들을 딕셔너리로 표현
print(soup.find("img").attrs["src"])
print(soup.find("img").attrs.get("src"))
print(soup.find("img").get("src"))
```

```python
# a 태그를 추출해주세요
print(soup.find("a"))
print(soup.find("a").attrs)
print(soup.find("a").attrs["href"])
print(soup.find("a").attrs.get("href"))
print(soup.find("a").get("href"))
```

### soup.select_one(경로)

```python
# 경로에 있는 첫번째 요소만 가져온다.
# 표현들: 태그.클래스이름, #id이름, 경로 연결은 ">"를 활용
# div안에 a 가 있는 요소를 추출하세요
print(soup.select_one("div > a"))
```

```python
# class 이름이 items인 div 안에 div를 추출하세요.
print(soup.select_one("div.items > div"))
```

```python
# id가 main2인 div를 추출하세요
print(soup.select_one("#main2"))
```

### soup.select(경로)

```python
# 경로에 있는 요소들을 가져온다.
# div안에 a 가 있는 요소를 추출하세요
print(soup.select("div > a"))
```

```python
# class 이름이 items인 div 안에 div를 추출하세요.
print(soup.select("div.items > div"))
```

## 1) HTML 요청해서 불러오기

```python
import requests

url = "https://search.kyobobook.co.kr/search?keyword=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5&gbCode=TOT&target=total"

response = requests.get(url)
print(response.status_code)
```

```python
# 누가 이 사이트에 요청을 했는가?
print(response.request.headers)
```

```python
# User-Agent 수동입력하기
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36"
}

url = "https://search.kyobobook.co.kr/search?keyword=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5&gbCode=TOT&target=total"

response = requests.get(url, headers=headers)
print(response.status_code)
```

```python
# 누가 이 사이트에 요청을 했는가?
print(response.request.headers)
```

```python
# 사용자 입력
keyword = "인공지능"
```

```python
# URL 만들기
url = "https://search.kyobobook.co.kr/search"

# "?keyword=%EC%9D%B8%EA%B3%B5%EC%A7%80%EB%8A%A5&target=kyobo&gbCode=TOT&page=2"  # & 로 이어진다.
params = {
    "keyword": keyword,  # 키워드
    "target": "kyobo",  # 교보문고 탭 선택
    "gbCode": "TOT",  # 통합검색
    "page": 1,  # 페이지 번호
}
```

```python
# 요청하기
response = requests.get(url, params=params, headers=headers)
print(response.status_code)
```

```python
print(response.request.url)
```

```python
# 웹에서 copy selector를 한 결과
# #shopData_list > ul > li:nth-child(1) > div.prod_area.horizontal > div.prod_thumb_box.size_lg

# 내가 적용해야 할 경로
# div.prod_area.horizontal > div.prod_thumb_box.size_lg
## 왜일까? 앞에 있는 경로 "#shopData_list > ul > li:nth-child(1)"는 prod_item이 그 결과이기 때문에
```

```python
left_box = prod_item.select_one(
    "div.prod_area.horizontal > div.prod_thumb_box.size_lg "
)
left_box
```

```python
right_box = prod_item.select_one("div.prod_area.horizontal > div.prod_info_box")
right_box
```

```python
title_box = right_box.select_one("div.auto_overflow_wrap.prod_name_group")
title_box.text
# 문자열 관련 메서드: strip(), replace(), split()
```

```python
title_list = title_box.text.strip().split("\n")
print(title_list)
```

```python
# 만약에 title_list 안에 "예약판매" 있으면 [0:3] 슬라이싱
# 아니면 [0:2] 슬라이싱
if "예약판매" in title_list:
    title = title_list[2]
else:
    title = title_list[1]

print(title)
```

```python
description = right_box.select_one("div.prod_desc_info").text.strip()
print(description)
```

```python
author_list = (
    right_box.select_one(
        "div.auto_overflow_wrap.prod_author_group > div.auto_overflow_contents > div"
    )
    .text.strip()
    .split()
)
print(author_list)
```

```python
publisher, publish_date = (
    right_box.select_one("div.prod_author_info > div.prod_publish")
    .text.strip()
    .split("\n·\n")
)
print(publisher)
print(publish_date)
```

```python
price_list = right_box.select_one("div.prod_price").text.strip().split()
discount_price = price_list[1]
price = price_list[4][:-1]  # '원' 제거
print(discount_price)
print(price)
```

```python
# 예약판매인 경우에는 없음 - 예외 처리
try:
    bottom_list = right_box.select_one("div.prod_bottom").text.strip().split()
    score = bottom_list[3]
    review_cnt = bottom_list[4][1:-1]  # 괄호 제거
except (AttributeError, IndexError):
    score = "없다"
    review_cnt = "없다"

print(score)
print(review_cnt)
```
