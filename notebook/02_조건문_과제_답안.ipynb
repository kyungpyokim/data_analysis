# 1. 편의점 영업: 9시부터 22시 사이면 "영업 중", 그 외에는 "영업 종료"를 출력하세요.


```python
# 변수 정의: 
current_hour = 14
# 조건: 
## 조건1: current_hour >= 9
## 조건2: current_hour <= 22
## 결합: AND
# 설계
if current_hour >= 9 and current_hour <= 22:
    print("영업 중")
else:
    print("영업 종료")
```

```python
# 변수 정의: 
my_score = 85
# 조건: 
## 조건1: my_score >= 90 (A)
## 조건2: my_score >= 80 (B)
## 조건3: my_score >= 70 (C)
## 결합: 순차적 비교(elif)
# 설계
if my_score >= 90:
    print("A")
elif my_score >= 80:
    print("B")
elif my_score >= 70:
    print("C")
else:
    print("F")
```

# 13. 당첨 확인: 당첨 번호들이 담긴 리스트에 내 번호가 있는지 확인하여 "1등 당첨"을 출력하세요.

```python
# 2개 이상 맞추면 1등 당첨
lottos = [1, 2, 3, 4, 5]
user = [1, 3, 5, 7, 6]

count = 0
for num in user:
    if num in lottos:
        count += 1
        print(num)
    if count == 2:
        print("1등입니다.")
        break
```

# 17. 잔액 확인: 원하는 물건이 상점 리스트에 있고, 내 지갑의 돈이 물건 가격보다 많아야 "구매 완료"를 출력하세요.


```python
# 변수 정의:
shop_items = {"우산": 5000}
my_money = 7000
pick = "우산"
# 조건: 
## 조건1: pick in shop_items
## 조건2: my_money >= shop_items[pick]
## 결합: AND
# 설계
if pick in shop_items and my_money >= shop_items[pick]:
    print("구매 완료")
else:
    print("구매 불가")
```

# 18. 가위바위보: 두 사람의 입력값을 비교하여 누가 이겼는지, 혹은 비겼는지 판정하세요.


```python
# 변수 
p1 = "바위"
p2 = "보"
print(f"Player1: {p1}, Player2: {p2}")

# p1이 이기는 딕셔너리 (p1:p2)
p1_win = {"바위": "가위", "가위":"보", "보":"바위"}

# 조건 1: p1 == p2 , 비김
# 조건 2: p1_win[p1] == p2, p1 이김
# 그 외 p2 이김
if p1 == p2:
    print("비김")
elif p1_win[p1] == p2:
    print("player1 이김") 
else:
    print("player2 이김")
```

```python
# 변수 정의: 
player1 = "가위"
player2 = "보"
# 조건: 
## 조건1: player1 == player2 (비김)
## 조건2: (player1 == "가위" and player2 == "보") or (player1 == "바위" and player2 == "가위") or (player1 == "보" and player2 == "바위") (p1 승리)
# 설계
if player1 == player2:
    print("비김")
elif (player1 == "가위" and player2 == "보") or (player1 == "바위" and player2 == "가위") or (player1 == "보" and player2 == "바위"):
    print("player1 승리")
else:
    print("player2 승리")
```

# 19. 스키장 개장: 현재 월(Month)이 겨울(12, 1, 2)에 속하고, 기온이 0도 이하일 때만 "스키장 개장"을 출력하세요.

# 20. 멤버십 혜택: 등급이 'VIP'이거나 'Gold'인 회원 중, 오늘 방문 횟수가 0인 경우에만 "무료 음료권 발급"을 출력하세요.

```python
# 변수 정의: 
grade = "VIP"
visit_count = 0
# 조건: 
## 조건1: grade in ["VIP", "Gold"]
## 조건2: visit_count == 0
## 결합: AND
# 설계
if grade in ["VIP", "Gold"] and visit_count == 0:
    print("무료 음료권 발급")
else:
    print("대상 아님")
```
