```python
import pandas as pd 

df1 = pd.read_csv(
    "./data/appreply2.csv",
    index_col=0
)

df2 = pd.read_csv(
    "./data/appreply2_한승헌.csv",
    index_col=0
)
```

```python
import pickle

with open("./data/word_list.pkl", "rb") as f:
    word_list1 = pickle.load(f)

print(word_list1)

import pickle

with open("./data/word_list_한승헌.pkl", "rb") as f:
    word_list2 = pickle.load(f)

print(word_list2)
```

```python
from konlpy.tag import Okt 
import re

okt = Okt()

review = df1["text"][36]
clean_sent = re.sub(r"[^0-9가-힣a-zA-Z\s]", "", review)
print(clean_sent)

result = okt.pos(clean_sent)
temp = []
for res in result:
    if res[1] ==  "Noun" and len(res[0]) > 1:
        temp.append(res[0])

print(len(temp))
print(temp)
```

```python
i = 0
for a, b in zip(word_list1, word_list2):
    if a != b:
        print(f"[{i}번째 리뷰]")
        print(f"강사: {len(a)} 비교: {len(b)}")
        print(f"강사: {set(a).symmetric_difference(set(b))}")
        print("="*50)
    i += 1
```
