# 데이터 설명
* customer_id: 고객 ID
* name: 고객 이름
* age: 고객 나이
* signup_date: 가입 날짜
* purchase_amount: 구매 금액
* is_active: 활성 상태

```python
import pandas as pd 

# 데이터 생성
data_dict = {
    "customer_id": [101, 102, 103, 104, 105],
    "name": ["Alice", "Bob", "Charlie", None, "Eva"],
    "age": [25.0, 30.0, 28.0, 40.0, 35.0],
    "signup_date": ["2023-01-15", "2023-02-20", "2023-03-20", "2023-03-01", "2023-04-10"],
    "purchase_amount": [120.50, 120.2, 200.00, 150.75, 300.40],
    "is_active": [True, False, "yes", "no", True],
    "group": ["A", "A", "B", "C", "C"]
}

# 데이터프레임 생성
df = pd.DataFrame(data_dict)
df
```

# 1. `is_active` 열만 추출(loc 이용, iloc 이용)


```python
# loc 이용
# 열 이름: is_active
df.loc[:, "is_active"]
```

```python
# iloc 이용
# 열 이름: is_active
# is_active는 앞에서 5번째 요소
df.iloc[:, -2] 
```

```python
# is_active는 뒤에서 2번째 요소
df.iloc[:, -2] 
```

# 2. `age`, `group` 열만 추출(loc 이용, iloc 이용)


```python
# loc 이용
# 열 이름: age, group 
df.loc[ : , ["age", "group"]]
```

```python
# iloc 이용
# 열 위치: age는 앞에서 2번째, group은 뒤에서 1번째
df.iloc[ : , [2, -1] ]
```

# 3. customer_id가 105인 행만 추출(loc 이용, iloc 이용)


```python
# loc 이용
# 행 이름: 4
df.loc[ 4 , :]
```

```python
# iloc 이용
# 행 위치: 앞에서 4번째
df.iloc[ 4 , :]
```

```python
# iloc 이용
# 행 위치: 뒤에서 -1번째
df.iloc[ -1 , :]
```

# 4. customer_id 가 103, 104, 105인 행만 추출(loc 이용, iloc 이용)


```python
# loc 이용
# 행 이름: 2, 3, 4
df.loc[ [2, 3, 4], :]
```

```python
# iloc 이용
# 행 번호: 앞에서 2번째부터 끝까지
df.iloc[ 2: , :]
```

```python
# iloc 이용
# 행 번호: 뒤에서 3번째부터 끝까지
df.iloc[ -3: , :]
```

# 6. customer_id 가 102, 103인 데이터의 name, age 추출(loc 이용, iloc 이용)


```python
# loc 이용
# 행 이름: 1, 2
# 열 이름: name, age
df.loc[ [1, 2], ["name", "age"] ]
```

```python
# iloc 이용
# 행 위치: 앞에서 1번째부터 3번째 앞까지
# 열 위치: 앞에서 1번째부터 3번째 앞까지
df.iloc[ 1:3, 1:3 ]
```
